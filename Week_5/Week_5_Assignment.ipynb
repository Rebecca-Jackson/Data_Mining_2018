```python
import numpy as np
from math import sqrt
M=matrix = np.loadtxt('prima.txt', usecols=range(9), skiprows=1)

numrows=len(M)
numcols=len(M[0])

def Mink(M,n,m):
    sum = 0
    for i in range(numrows):
        sum += ((abs(M[i][n]-M[i][m]))**3)
    return ((sum)**(1/3))
for i in range(numcols):
    print ('The Minkowski Distance between column 3 and column ', i+1 ,' is ', Mink(M,2,i)) 
```
