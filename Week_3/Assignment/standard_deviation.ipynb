```python
import math
import numpy as np
D=[[34,43,35,53,36,63,37],
[38,83,39,93,40,4,41],
[42,24,43,34,44,44,45],
[45,54,46,64,47,74,48],
[49,94,50,5,51,15,52],
[53,35,54,45,55,55,56],
[57,75,58,85,59,95,60]]

def sum_row(M,n):
    sum=0
    for i in range(7):
        sum+= M[n][i]
    return sum

def sum_column(M,n):
    sum=0
    for i in range(7):
        sum+= M[i][n]
    return sum

def sum_matrix(M):
    sum=0
    for i in range(0,6):
        sum+=sum_column(M,i)
    return sum

def mean_row(M,n):
    return sum_row(M,n)/len(M)

def mean_column(M,n):
    col=np.size(M)/len(M)
    return sum_column(M,n)/col

def mean_matrix(M):
    return sum_matrix(M)/np.size(M)

def sd_row(M,n):
    x=mean_row(M,n)
    sum=0
    for i in range(0,6):
        sum+= math.pow(2,M[i][n]-x)
    return math.sqrt(sum/len(M))

def sd_column(M,n):
    col=np.size(M)/len(M)
    x=mean_column(M,n)
    sum=0
    for i in range(0,6):
        sum+= math.pow(2,M[n][i]-x)
    return math.sqrt(sum/col)

def sd_matrix(M):
    
    x=mean_matrix(M)
    sum=0
    for i in range(0,6):
        for j in range(0,6):
            sum+= math.pow(2,M[i][j]-x)
    return math.sqrt(sum/np.size(M))

print(sd_matrix(D))
print(sd_column(D,1))
```
